AIM - ANOVA (Analysis of Variance)

 Perform one-way ANOVA to compare means across multiple groups.

 Conduct post-hoc tests to identify significant differences between group means.

### **ANOVA (F-TEST)**

**One Way F-test (Anova):** We will compare the means of **Balance** across different **Education** levels (primary, secondary, tertiary, unknown) to see if they are similar.

```python
import pandas as pd
import scipy.stats as stats

# Load the dataset
df = pd.read_csv("bank.csv")

# Sample data: We will use the 'age' column
data = df['age']

# Prepare data for ANOVA
d_data = {grp: df['balance'][df.education == grp] for grp in df.education.unique()}

# Perform One-way ANOVA
F, p = stats.f_oneway(d_data['secondary'], d_data['tertiary'], d_data['primary'], d_data['unknown'])

print("p-value for significance is: ", p)

if p < 0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

### **Two Way F-test**

We will check the effect of two independent variables (**Education** and **Marital**) on a dependent variable (**Balance**).

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Fit the OLS model
# We are testing: Balance ~ Education + Marital + Education*Marital
model = ols('balance ~ C(education)*C(marital)', df).fit()

print(f"Overall model F({model.df_model: .0f},{model.df_resid: .0f}) = {model.fvalue: .3f}, p = {model.f_pvalue: .4f}")

# Perform ANOVA on the model
res = sm.stats.anova_lm(model, typ=2)
print(res)
```
