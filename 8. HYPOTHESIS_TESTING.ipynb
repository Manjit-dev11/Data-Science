### t-test to evaluate whether our hypothesis is correct or not.

```
import pandas as pd
import scipy.stats as stats

# Load the dataset
df = pd.read_csv("bank.csv")

# Sample data: We will use the 'age' column
data = df['age']

print("First 10 Ages : ", data.head(10).values)

# Define the null hypothesis
H0 = "The average Age of clients is 40."

# Define the alternative hypothesis
H1 = "The average Age of clients is not 40."

# Calculate the test statistic
# We perform a 1-sample t-test against the population mean of 40
t_stat, p_value = stats.ttest_1samp(data, 40)

# Print the results
print("Test statistic:", t_stat)
print("p-value:", p_value)

# Conclusion
if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

### **Two sampled T-test**

The Independent Samples t-test compares the means of two independent groups. Here, we will compare the **Balance** of clients who subscribed to a term deposit (**deposit = yes**) vs those who did not (**deposit = no**) to see if there is a significant difference.

```
from scipy.stats import ttest_ind
import numpy as np

# Select two independent groups
# We compare the Balance of clients who made a deposit vs those who didn't
deposit_yes_balance = df[df['deposit'] == 'yes']['balance']
deposit_no_balance = df[df['deposit'] == 'no']['balance']

print("Deposit=Yes Balance (first 5):", deposit_yes_balance.head().values)
print("Deposit=No Balance (first 5):", deposit_no_balance.head().values)

yes_mean = np.mean(deposit_yes_balance)
no_mean = np.mean(deposit_no_balance)

print("Deposit=Yes mean balance:", yes_mean)
print("Deposit=No mean balance:", no_mean)

yes_std = np.std(deposit_yes_balance)
no_std = np.std(deposit_no_balance)

print("Deposit=Yes std deviation:", yes_std)
print("Deposit=No std deviation:", no_std)

# Perform Independent t-test
ttest, pval = ttest_ind(deposit_yes_balance, deposit_no_balance)
print("p-value", pval)

if pval < 0.05:
    print("We reject null hypothesis")
else:
    print("We accept null hypothesis")
```

### **Paired sampled t-test**

The paired sample t-test checks for differences between two related variables. Since this dataset is a snapshot (not time-series), we will create a synthetic "Previous Balance" for demonstration purposes.

**H0**: Mean difference between current and previous balance is 0.
**H1**: Mean difference is not 0.

```
import pandas as pd
from scipy import stats
import numpy as np

# Simulating a "Previous Balance" by adding random noise to the current balance
# This is for demonstration as the dataset doesn't have paired time-series data
np.random.seed(42)
df['balance_prev'] = df['balance'] + np.random.randint(-100, 100, size=len(df))

# Selecting a subset for clearer demonstration
subset_df = df.head(100)

print(subset_df[['balance_prev', 'balance']].describe())

# Perform Paired t-test
ttest, pval = stats.ttest_rel(subset_df['balance_prev'], subset_df['balance'])
print("p-value:", pval)

if pval < 0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

### **When you can run a Z Test.**

We use a Z-test because our sample size is large (N \> 30). We will test if the mean **Balance** is significantly different from 1500.

**One-sample Z test**

```
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests

# H0: The mean Balance is 1500
ztest, pval = stests.ztest(df['balance'], x2=None, value=1500)

print(float(pval))

if pval < 0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

### **Two-sample Z test**

Here we check two independent data groups. We will compare the **Balance** of clients with a **Housing Loan** vs clients **without a Housing Loan**.

**H0**: Mean Balance of Housing Loan (yes) and Housing Loan (no) groups is equal.
**H1**: Mean Balance is not equal.

```
# Selecting two independent groups
housing_yes_balance = df[df['housing'] == 'yes']['balance']
housing_no_balance = df[df['housing'] == 'no']['balance']

# Perform Two-sample Z-test
ztest, pval1 = stests.ztest(housing_yes_balance, x2=housing_no_balance,
                            value=0, alternative='two-sided')

print(float(pval1))

if pval1 < 0.05:
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

### **Chi-Square Test**

This test is applied to two categorical variables. We will determine if there is a significant association between **Marital Status** and **Deposit** (subscription status).

```
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df['marital'], df['deposit'])
print("Contingency Table:\n", contingency_table)

# Perform the chi-square test
chi2_statistic, p_value, dof, expected_frequencies = chi2_contingency(contingency_table)

# Print the results
print('Chi-square statistic:', chi2_statistic)
print('P-value:', p_value)
print('Degrees of freedom:', dof)
print('Expected frequencies:')
print(expected_frequencies)

if p_value < 0.05:
    print("Reject null hypothesis (Variables are related)")
else:
    print("Accept null hypothesis (Variables are independent)")
```
